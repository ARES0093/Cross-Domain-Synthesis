```python
import pandas as pd
from imblearn.over_sampling import SMOTE


df = pd.read_csv('../data/raw/CICIDS2017/cicids2017_cleaned.csv')

X = df.drop(columns=['Attack Type'])
y = df['Attack Type']

print("Data loaded and split successfully!")
```

```python

y.value_counts()
```

```python

strategy = {
    'DoS': 200000,
    'DDoS': 200000,
    'Port Scanning': 200000,
    'Brute Force': 200000,
    'Web Attacks': 200000,
    'Bots': 200000
}


smote = SMOTE(sampling_strategy=strategy, random_state=42)


print("Synthesizing new attack data... Please wait.")
X_resampled, y_resampled = smote.fit_resample(X, y)


y_resampled.value_counts()
```

```python

df_balanced = X_resampled.copy()
df_balanced['Attack Type'] = y_resampled


df_balanced.to_csv('../data/processed/cicids2017_balanced.csv', index=False)

print("SMOTE balancing complete and saved to disk!")
```
